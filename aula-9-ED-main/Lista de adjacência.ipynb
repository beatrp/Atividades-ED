```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo
G = nx.Graph()

# Adicionar vértice / nós ()
G.add_nodes_from(['Audrey', 'Bill', 'Cathy', 'Dave', 'Emily', 'Frank', 'Gayle', 'Harry', 'Hana', 'Jeff'])
```

```python
G.add_edges_from([
    ('Audrey', 'Bill'),
    ('Audrey', 'Gayle'),
    ('Audrey', 'Hana'),
    ('Bill', 'Emily'),
    ('Bill', 'Jeff'),
    ('Bill', 'Gayle'),
    ('Cathy', 'Emily'),
    ('Cathy', 'Gayle'),
    ('Dave', 'Emily'),
    ('Dave', 'Hana'),
    ('Dave', 'Frank'),
    ('Emily', 'Jeff'),
    ('Emily', 'Frank'),
    ('Harry', 'Jeff'),
    ('Harry', 'Hana'),])
```

```python
print(f"Nós do grafo: {G.nodes()}")
print(f"Arestas do grafro: {G.edges()}")
print(f"Contando o num de Nós: {G.number_of_nodes()}")
print(f"Contando o num de Arestas: {G.number_of_edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels= True, node_color='lightgreen',
        font_weight='bold',node_size=700, font_size=10)

plt.title('Representação em Grafo simples')
plt.show()
```
