```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo orientado
G = nx.DiGraph()

# Adicionar vértices
G.add_nodes_from(['Beyonce', 'David Bowie', 'Jay-Z', 'Kanye West', 'Lady Gaga', 'Madonna', 'Michael Jackson'])
```

```python
G.add_edges_from([
    ('Beyonce', 'Rihanna'),
    ('David Bowie','Madonna'),
    ('David Bowie','Prince'),
    ('Jay-Z','Kanye West'),
    ('Jay-Z','Rihanna'),
    ('Madonna','Lady Gaga'),
    ('Michael Jackson','Beyonce'),
    ('Michael Jackson','Rihana'),
    ('Prince','Michael Jackson'),])
```

```python
print(f"Nós do grafo: {G.nodes()}")
print(f"Arestas do grafro: {G.edges()}")
print(f"Contando o num de Nós: {G.number_of_nodes()}")
print(f"Contando o num de Arestas: {G.number_of_edges()}")
```

```python
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels= True, node_color='red',
        node_size=2500, font_color='black', font_size=8, arrowsize=15)

plt.title('Representação em Grafo simples')
plt.show()
```
